```python
#Step 1 : import library

import pandas as pd

#Step 2 : import data

df = pd.read_csv("https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Boston.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Step 3 :Define y and x
df.columns
```

```python
y = df['MEDV']
X = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
#Step 4 : Train test split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test =train_test_split(X,y,train_size=0.7,random_state=2595)
```

```python
#Check shape of train and test sample
X_train.shape,
X_test.shape,
y_train.shape,
y_test.shape
```

```python
#Step 5 : Select model
from sklearn.linear_model import LinearRegression
model = LinearRegression
```

```python
#Step 6 : train or fit model

model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#step 7 : Predict Model
y_predict=model.predict(X_test)
```

```python
#Step 8 :Model Accuracy
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_predict)
```

```python
intercept = model.intercept_
coefficients = model.coef_
```

```python
print("Intercept:", intercept)
print("Coefficients:", coefficients)
print("Mean Absolute Percentage Error:", mean_absolute_percentage_error)
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of actual vs. predicted values
plt.scatter(y_test, y_predict)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs. Predicted Values")
plt.show()
```

```python
# Create a residual plot
residuals = y_test - y_predict
plt.scatter(y_predict, residuals)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.axhline(y=0, color='r', linestyle='--')
plt.title("Residual Plot")
plt.show()
```

```python
# Create a bar chart of feature coefficients
plt.bar(X.columns, model.coef_)
plt.xlabel("Features")
plt.ylabel("Coefficient Value")
plt.title("Feature Coefficients")
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a scatter plot with the regression line
plt.scatter(X_test['RM'], y_test, label="Actual Values")
plt.plot(X_test['RM'], y_predict, color='red', linewidth=2, label="Regression Line")
plt.xlabel("RM (Number of Rooms)")
plt.ylabel("MEDV")
plt.title("Regression Line Plot")
plt.legend()
plt.show()
```
